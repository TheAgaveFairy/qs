```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("results2.csv")
df['Log2MatrixSize'] = np.log2(df['MatrixSize'])
df
```

```python
grouped = df.groupby(['Threads', 'MatrixSize']).agg({
    'Time': ['mean', 'std']
}).reset_index()
grouped['Log2MatrixSize'] = np.log2(grouped['MatrixSize'])
grouped['Log2MatrixSize'] = grouped['Log2MatrixSize'].astype(int)
grouped.columns = ['Threads', 'MatrixSize', 'MeanTime', 'StdTime', 'Log2MatrixSize']
grouped = grouped[grouped['Log2MatrixSize'] > 4]
```

```python
grouped.to_csv('table.csv')
```

```python
most = max(grouped['Log2MatrixSize'])

plt.figure(dpi=1200)

for i in range(1, most//2 + 1):
    temp = grouped[grouped['Log2MatrixSize'] == i]
    #print(temp)
    plt.errorbar(temp['Threads'], temp['MeanTime'], yerr=temp['StdTime'], label=f"n = 2^{i}")
    plt.xlabel("Threads")
    plt.ylabel("Mean Time")
plt.title(f"Mean Times for Mat. Mul. w/ stddev Pt. 1")
plt.legend(loc='upper right')
plt.savefig('results1.png')
plt.show()
```

```python
plt.figure(dpi=1200)

for i in range(most//2 + 1, most + 1):
    temp = grouped[grouped['Log2MatrixSize'] == i]
    #print(temp)
    plt.errorbar(temp['Threads'], temp['MeanTime'], yerr=temp['StdTime'], label=f"n = 2^{i}")
    plt.xlabel("Threads")
    plt.ylabel("Mean Time")
plt.title(f"Mean Times for Mat. Mul. w/ stddev Pt. 2")
plt.legend(loc='upper right')
plt.savefig('results2.png')
plt.show()
```
